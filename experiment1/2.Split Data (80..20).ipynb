```python
# 1. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# 2. Get Data
df = pd.read_csv('train.csv')

# 3. Split Data into Train and Test sets
train_df, test_df = train_test_split(df, test_size=0.2, stratify=df['score'], random_state=42)

# 4. Save Train and Test Data Sets
train_df.to_csv('train_split.csv', index=False)
test_df.to_csv('test_split.csv', index=False)

# 5. Plot Score Distribution for Train and Test sets
train_score_counts = train_df['score'].value_counts().sort_index()
test_score_counts = test_df['score'].value_counts().sort_index()

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# Train Data Set
train_score_counts.plot(kind='bar', ax=axes[0], color='skyblue')
axes[0].set_title('Distribution of Scores in Train Data Set')
axes[0].set_xlabel('Score')
axes[0].set_ylabel('Frequency')

# Test Data Set
test_score_counts.plot(kind='bar', ax=axes[1], color='salmon')
axes[1].set_title('Distribution of Scores in Test Data Set')
axes[1].set_xlabel('Score')
axes[1].set_ylabel('Frequency')

plt.tight_layout()

# Save the plot
plt.savefig('score_distribution_train_test_plot.png')

# Show the plot
plt.show()
```
